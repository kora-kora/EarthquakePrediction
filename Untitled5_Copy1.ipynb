```python
#load data
nny = 300
nnx = 300 
velocity = np.ones((nnx,nny))
velocity[0:100,:] = 3000.
velocity[100:200,:] = 5000.
velocity[200:300,:] = 6000.
```

```python
#define velocity density and spacing
dx = 10
density = 1500

for1   = forward(velocity,density,dx)

#call FDpar() to show FD parameter recomendation
for1.FDpar()
```

```python
#define ricker source based on the above parameters 
# t is the time axis ; 
# s is the amplitude ;
# so you can plot plt.plot(t,s) or put plot = true

t,s= sources(dt = 0.0005,nt = 2000, fc = 15.).ricker(a=4,plot=True)
```

```python
#add PML to the velocity
npml = 100
cpml   = for1.pml(npml)

#apply absorbing boundary condition to the PML
for1.ApplyPML(1,1)
```

```python
#geometry 
sx = 50
sy = 40

#set receiver location
rex,rez   = np.linspace(150,190,41,dtype=int),np.linspace(50,90,41,dtype=int)
recx,recz = np.meshgrid(rex,rez); recx = np.ravel(recx); recz = np.ravel(recz)
```

```python
#plot setup
plt.figure(num=None,figsize=(6,5), dpi=500, facecolor='w', edgecolor='k')
plt.style.use('seaborn-paper')
plt.imshow(velocity,cmap="RdBu_r",extent=[0,(nnx*dx)-dx,(nny*dx)-dx,0])
plt.colorbar(fraction=0.02, pad=0.06,shrink = 0.4 , orientation="vertical",label="velocity [m/s]")
plt.plot(sx*dx,sy*dx,'r*')
plt.plot(recx*dx,recz*dx,'gv',markersize = 2)
plt.xlabel('x [m]')
plt.ylabel('z [m]')
plt.title('Velocity model')
plt.rcParams.update({'font.size': 6})
```

```python
#run the setup
#to be able to show the movie add '%matplotlib qt' befor the solve module and set plotmov to true
#to make plotted image in the jupyter notebook again add '%matplotlib inline'
%matplotlib qt 

res = for1.solve(recx,recz,sx,sy,t,s,plotmov=True)

%matplotlib inline
```

```python
list(res)
```

```python
import scipy.integrate as it

ux = it.cumtrapz(res['Vx'][:,50], x=t)
uy = it.cumtrapz(res['Vy'][:,50], x=t)

uxx = np.sqrt(ux**2+uy**2)
uxxx = np.gradient(uxx,0.5)
```
